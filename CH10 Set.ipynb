# Set
  - Sets are unordered.
  - Set elements are unique. 
  - Duplicate elements are not allowed.
  - A set itself may be modified, but the elements contained in the set must be of an immutable type

## CRUD - Update - Modifying a set 

```
# initialize my_set
myset = {1, 3}
print(myset)

# if you uncomment line 9,
# you will get an error
# TypeError: 'set' object does not support indexing

# myset[0]

# add an element
# Output: {1, 2, 3}
myset.add(2)
print(myset)

# add multiple elements
# Output: {1, 2, 3, 4}
myset.update([2, 3, 4])
print(myset)

# add list and set
# Output: {1, 2, 3, 4, 5, 6, 8}
myset.update([4, 5], {1, 6, 8})
print(myset)
```

## CRUD - Delete - Remove Set Items

```
# Difference between discard() and remove()

# initialize myset
myset = {1, 3, 4, 5, 6}
print(myset)

# discard an element
# Output: {1, 3, 5, 6}
myset.discard(4)
print(myset)

# remove an element
# Output: {1, 3, 5}
myset.remove(6)
print(myset)

# discard an element
# not present in myset
# Output: {1, 3, 5}
myset.discard(2)
print(myset)

# remove an element
# not present in myset
# you will get an error.
# Output: KeyError

myset.remove(2)
```

```
# initialize myset
# Output: set of unique elements
myset = set("HelloWorld")
print(myset)

# pop an element
# Output: random element
print(myset.pop())

# pop another element
myset.pop()
print(myset)

# clear my_set
# Output: set()
myset.clear()
print(myset)
```

```
#@title Intersection
```

```
computer = {'Khan','Lisa','lina','Alisa','Emma'}
math = {'Alisa','Rean','Raura','Emma','Jusis'}
both = computer & math

print(f'computer : {computer}')
print(f'math : {math}')
print(f'intersection : {both}')
```

```
computer = {'Khan','Lisa','lina','Alisa','Emma'}
math = {'Alisa','Rean','Raura','Emma','Jusis'}
both = computer.intersection(math)

print(f'computer : {computer}')
print(f'math : {math}')
print(f'intersection : {both}')
```

```
#@title Union
```

```
computer = {'Khan','Lisa','lina','Alisa','Emma'}
math = {'Alisa','Rean','Raura','Emma','Jusis'}
both = computer | math

print(f'computer : {computer}')
print(f'math : {math}')
print(f'union : {both}')
```

```
computer = {'Khan','Lisa','lina','Alisa','Emma'}
math = {'Alisa','Rean','Raura','Emma','Jusis'}
both = computer.union(math)

print(f'computer : {computer}')
print(f'math : {math}')
print(f'union : {both}')
```

```
#@title Difference
```

```
computer = {'Khan','Lisa','lina','Alisa','Emma'}
math = {'Alisa','Rean','Raura','Emma','Jusis'}
both = computer - math
both2 = math - computer

print(f'computer : {computer}')
print(f'math : {math}')
print(f'difference computer - math: {both}')
print(f'difference math - computer : {both2}')
```

```
computer = {'Khan','Lisa','lina','Alisa','Emma'}
math = {'Alisa','Rean','Raura','Emma','Jusis'}
both = computer.difference(math)
both2 = math.difference(computer)

print(f'computer : {computer}')
print(f'math : {math}')
print(f'difference computer - math: {both}')
print(f'difference math - computer : {both2}')
```

```
#@title symmetric difference
```

```
computer = {'Khan','Lisa','lina','Alisa','Emma'}
math = {'Alisa','Rean','Raura','Emma','Jusis'}
both = computer ^ math
both2 = math ^ computer

print(f'computer : {computer}')
print(f'math : {math}')
print(f'symmetric difference computer ^ math: {both}')
print(f'symmetric difference math ^ computer: {both2}')
```

```
computer = {1,2,3,4,5}
math = {3,4,5,6,7}
both = computer.symmetric_difference(math)
both2 = math.symmetric_difference(computer)

print(f'computer : {computer}')
print(f'math : {math}')
print(f'symmetric difference computer ^ math: {both}')
print(f'symmetric difference math ^ computer: {both2}')
```

## Set Method

```
#@title add()
Adds an element to the set
```

```
fruits = {"apple", "banana", "cherry"}
print(f'Before : {fruits}')
fruits.add("orange")
print(f'After : {fruits}')
```

```
#@title copy()
Returns a copy of the set
```

```
fruits = {"apple", "banana", "cherry"}
print(f'fruits : {fruits}')
print(f'address : {id(fruits)}')

cp = fruits.copy()
print(f'copy : {cp}')
print(f'address : {id(cp)}')
```

```
#@title discard()
Remove the specified item
```

```
fruits = {"apple", "banana", "cherry"}
print(f'orign fruits:{fruits}')
fruits.discard("banana")
print(f'remove fruits:{fruits}')
```

```
#@title pop()
Removes an element from the set
```

```
fruits = {"apple", "banana", "cherry"}
print(f'orign fruits:{fruits}')
fruits.pop()
print(f'pop fruits:{fruits}')
```

```
#@title clear()
Removes all the elements from the set
```

```
fruits = {"apple", "banana", "cherry"}
print(f'orign fruits:{fruits}')
fruits.clear()
print(f'clear fruits:{fruits}')
```

```
#@title isdisjoint()
returns True if none of the items are present in both sets, otherwise it returns False.
```

```
x = {"apple", "banana", "cherry"}
y = {"google", "microsoft", "facebook"}
z = x.isdisjoint(y)

print(f'x:{x}')
print(f'y:{y}')
print(f'z:{z}')
```

```
#@title issubset()
Returns whether another set contains this set or not
```

```
x = {"a", "b", "c"}
y = {"f", "e", "d", "c", "b", "a"}
z = x.issubset(y)

print(f'x:{x}')
print(f'y:{y}')
print(f'z:{z}')
```

```
#@title issuperset()
Returns whether this set contains another set or not
```

```
x = {"f", "e", "d", "c", "b", "a"}
y = {"a", "b", "c"}
z = x.issuperset(y)

print(f'x:{x}')
print(f'y:{y}')
print(f'z:{z}')
```

```
#@title update()
Update the set with the union of this set and others
```

```
x = {"apple", "banana", "cherry"}
print(f'x:{x}')

y = {"google", "microsoft", "apple"}
print(f'y:{y}')

x.update(y)
print(f'update : {x}')
```

```
#@title intersection_update()
Removes the items in this set that are not present in other, specified set(s)
```

```
x = {"apple", "banana", "cherry"}
print(f'x:{x}')

y = {"google", "microsoft", "apple"}
print(f'y:{y}')

x.intersection_update(y)
print(f'update : {x}')
```

```
#@title difference_update()
Removes the items in this set that are also included in another, specified set
```

```
x = {"apple", "banana", "cherry"}
print(f'x:{x}')

y = {"google", "microsoft", "apple"}
print(f'y:{y}')

x.difference_update(y)
print(f'update : {x}')
```

```
#@title symmetric_difference_update()
inserts the symmetric differences from this set and another
```

```
x = {"apple", "banana", "cherry"}
print(f'x:{x}')

y = {"google", "microsoft", "apple"}
print(f'y:{y}')

x.symmetric_difference_update(y)
print(f'update : {x}')
```

## Frozenset
  - Frozenset is a new class that has the characteristics of a set, but its elements cannot be changed once assigned

```
A = frozenset([1, 2, 3, 4])
B = frozenset([3, 4, 5, 6])
 
print(f'A : {A}')
print(f'B : {B}')

print(f'Union : {A | B}')
print(f'Intersection : {A & B}')
```
