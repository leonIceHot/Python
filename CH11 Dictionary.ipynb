# Dictionary
  - Unordered
  - Changeable

```
#@title Definition
```

```
age = {'Rean':18,'Alisa':17,'Laura':17,'Machisa':17,
       'Jusis':17,'Emma':17,'Fie':15,'Gaius':17}

print(f'age:{age}')
```

```
#@title Dictionary Items
```

```
age = {'Rean':18,'Alisa':17,'Laura':17,'Machisa':17,
       'Jusis':17,'Emma':17,'Fie':15,'Gaius':17}

print(f'Rean age : {age["Rean"]}')
print(f'Alisa age : {age["Alisa"]}')
print(f'Laura age : {age["Laura"]}')
print(f'Machisa age : {age["Machisa"]}')
print(f'Jusis age : {age["Jusis"]}')
print(f'Emma age : {age["Emma"]}')
print(f'Fie age : {age["Fie"]}')
print(f'Gaius age : {age["Gaius"]}')
```

```
#@title Duplicates Not Allowed
```

```
# Dictionaries cannot have two items with the same key:
```

```
age = {'Rean':18,'Alisa':17,'Laura':17,'Machisa':17,
       'Jusis':17,'Emma':17,'Fie':15,'Gaius':17,
       'Rean':20}
print(f'age : {age}')
```

```
#@title Dictionary Length
```

```
# use the len() function
```

```
age = {'Rean':18,'Alisa':17,'Laura':17,'Machisa':17,
       'Jusis':17,'Emma':17,'Fie':15,'Gaius':17,
       'Rean':20}
print(f'length : {len(age)}')
```

```
#@title Dictionary Items - Data Types
```

```
# The values in dictionary items can be of any data type:
```

```
age = {'Rean':18,'Alisa':False,'Laura':'17','Machisa':17,
       'Jusis':17,'Emma':['COLOR'],'Fie':15,'Gaius':17,
       'Rean':20}
print(f'Item type : {age}')
```

```
#@title Dictionary type
```

```
# From Python's perspective, dictionaries are defined as objects with the data type 'dict':
```

```
age = {'Rean':18,'Alisa':17,'Laura':17,'Machisa':17,
       'Jusis':17,'Emma':17,'Fie':15,'Gaius':17}
print(f'age type : {type(age)}')
```

```
#@title Accessing Items
```

```
age = {'Rean':18,'Alisa':17,'Laura':17,'Machisa':17,
       'Jusis':17,'Emma':17,'Fie':15,'Gaius':17}

print(f'age keys : {age.keys()}')
print(f'age values : {age.values()}')
print(f'age items : {age.items()}')
```

```
#@title Check if Key Exists
```

```
age = {'Rean':18,'Alisa':17,'Laura':17,'Machisa':17,
       'Jusis':17,'Emma':17,'Fie':15,'Gaius':17}

print(f'age: {age}')

if "Rean" in age:
    print("Rean age is 18!")
```

```
#@title Change Values
```

```
age = {'Rean':18,'Alisa':17,'Laura':17,'Machisa':17,
       'Jusis':17,'Emma':17,'Fie':15,'Gaius':17}

print(f'Fie age: {age["Fie"]}')
age["Fie"]=20
print(f'Fie age: {age["Fie"]}')
```

```
age = {'Rean':18,'Alisa':17,'Laura':17,'Machisa':17,
       'Jusis':17,'Emma':17,'Fie':15,'Gaius':17}

print(f'Fie age: {age["Fie"]}')
age.update({"Fie":20})
print(f'Fie age: {age["Fie"]}')
```

```
#@title Adding Items
```

```
age = {'Rean':18,'Alisa':17,'Laura':17,'Machisa':17,
       'Jusis':17,'Emma':17,'Fie':15,'Gaius':17}

print(f'age:{age}')
age['Angelica']=20
print(f'age:{age}')
```

```
age = {'Rean':18,'Alisa':17,'Laura':17,'Machisa':17,
       'Jusis':17,'Emma':17,'Fie':15,'Gaius':17}

print(f'age:{age}')
age.update({'Angelica':20})
print(f'age:{age}')
```

```
#@title Removing Items
```

```
age = {'Rean':18,'Alisa':17,'Laura':17,'Machisa':17,
       'Jusis':17,'Emma':17,'Fie':15,'Gaius':17}

print(f'age:{age}')
age.pop("Rean")
print(f'age:{age}')
```

```
age = {'Rean':18,'Alisa':17,'Laura':17,'Machisa':17,
       'Jusis':17,'Emma':17,'Fie':15,'Gaius':17}

print(f'age:{age}')
age.popitem()
print(f'age:{age}')
```

```
age = {'Rean':18,'Alisa':17,'Laura':17,'Machisa':17,
       'Jusis':17,'Emma':17,'Fie':15,'Gaius':17}

print(f'age:{age}')
del age['Rean']
print(f'age:{age}')
```

```
age = {'Rean':18,'Alisa':17,'Laura':17,'Machisa':17,
       'Jusis':17,'Emma':17,'Fie':15,'Gaius':17}

print(f'age:{age}')
age.clear()
print(f'age:{age}')
```

```
#@title Loop Dictionaries
```

```
age = {'Rean':18,'Alisa':17,'Laura':17,'Machisa':17,
       'Jusis':17,'Emma':17,'Fie':15,'Gaius':17}

for x in age:
    print(x)
```

```
age = {'Rean':18,'Alisa':17,'Laura':17,'Machisa':17,
       'Jusis':17,'Emma':17,'Fie':15,'Gaius':17}
       
for x in age.keys():
    print(x)
```

```
age = {'Rean':18,'Alisa':17,'Laura':17,'Machisa':17,
       'Jusis':17,'Emma':17,'Fie':15,'Gaius':17}
       
for x in age.values():
    print(x)
```

```
age = {'Rean':18,'Alisa':17,'Laura':17,'Machisa':17,
       'Jusis':17,'Emma':17,'Fie':15,'Gaius':17}

for x,y in age.items():
    print(x,y)
```

```
#@title Nested Dictionaries
```

```
myfamily = {
  "child1" : {
    "name" : "Emil",
    "year" : 2004
  },
  "child2" : {
    "name" : "Tobias",
    "year" : 2007
  },
  "child3" : {
    "name" : "Linus",
    "year" : 2011
  }
}

print(f'myfamily:{myfamily}')
```
