```
x = ["apple", "banana"]

print("pineapple" not in x)

# returns True because a sequence with the value "pineapple" is not in the list
```

## Any and All

```
def all(iterable):
    for element in iterable:
        if not element:
            return False
    return True
```

```
# all
nums = [1,1,0,1]
print(all(nums))

chars = ['a','b','c','d']
print(all(chars))
```

```
def any(iterable):
    for element in iterable:
        if element:
            return True
    return False
```

```
# any
nums = [1,1,0,1]
print(any(nums))

chars = ['a','b','c','d']
print(any(chars))
```

## Copy Everything with deepcopy()
  - =賦值： 
           對於修改，受到影響，位址相同
  - 淺複製 shallow copy：
           對於不變型別的修改，受影響；
           對於可變型別的修改，不受影響，位址不相同
  - 深複製 deep copy：
           對於不變型別的修改，不受影響；
           對於可變型別的修改，不受影響，位址不相同

```
a = [1,2]
b = a
print(f'a:{a,id(a)}')
print(f'b:{b,id(b)}')
```

```
import copy

a = [1,[2,3]]
equal = a
shallow = copy.copy(a)
deepcopy = copy.deepcopy(a)

print(f'a:{a,id(a)}')
print(f'equal:{equal,id(equal)}')
print(f'shallow:{shallow,id(shallow)}')
print(f'deepcopy:{deepcopy,id(deepcopy)}')
```

```
import copy

a = [1,[2,3]]
equal = a
shallow = copy.copy(a)
deepcopy = copy.deepcopy(a)

a[0] = 4
a[1][1] = 5

print(f'a:{a,id(a)}')
print(f'equal:{equal,id(equal)}')
print(f'shallow:{shallow,id(shallow)}')
print(f'deepcopy:{deepcopy,id(deepcopy)}')
```

## Iterate Multiple Sequences with zip()

```
# 第一個 List
names = ["A", "B", "C"]

# 第二個 List
values = [11, 23, 46]

# 使用 zip 同時迭代兩個 List
for x, y in zip(names, values):
  print(x, y)
```

```
# 兩個不同長度的 List
names2 = ["A", "B", "C"]
values2 = [11, 23]
for x, y in zip(names2, values2):
  print(x, y)
```

```
names = ["A", "B", "C"]
values = [11, 23, 46]

# 打包成 Tuples 組成的 List
zippedList = list(zip(names, values))
print(zippedList)

# 轉回原來的兩個 Lists
n, v = zip(*zippedList)
print(n)
print(v)    
```

## Create a List with a Comprehension
  - [ expression for element in iterable ]
  - [ expression for element in iterable if condition]

```
#@title [expression for element in iterable]
```

```
squares = []
for x in (1,2,3,4):
    squares.append(x*x)
print(squares)
```

```
squares = [x * x for x in (1,2,3,4)]
print(squares)
```

```
for x in (1,2,3):
  print(x)
```

```
[print(x) for x in (1,2,3)]
```

```
#@title [expression for element in iterable if condition]
```

```
[print(x) for x in range(10) if x % 2 == 0]
```

```
#@title Whitespace in list comprehensions
```

```
squares = [x * x for x in (1,2,3,4)]
print(squares)
```

```
square=[
 x * x 
 for x in (1,2,3,4)
]
print(square)
```

```
#@title List Comprehensions with Nested Loops
```

```
data = [[1,2],[3,4],[5,6]]
output = []
for each_list in data:
    for element in each_list:
        output.append(element)
print(output)
```

```
data = [[1,2],[3,4],[5,6]]
output =[element for each_list in  data for element in each_list]
print(output)
```
