  - Python has two primitive loop commands:

    - for loops   -- 是一個有限的循環

## Operators
 - Python Arithmetic Operators
  - Arithmetic operators are used with numeric values to perform common mathematical operations 

 - Python Assignment Operators
  - Assignment operators are used to assign values to variables

```
#@title Python Arithmetic Operators
 
```

```
# Arithmetic Operators
x = 15
y = 4

# Output: x + y = 19
print('x + y =',x+y)

# Output: x - y = 11
print('x - y =',x-y)

# Output: x * y = 60
print('x * y =',x*y)

# Output: x / y = 3.75
print('x / y =',x/y)

# Output: x % y = 3
print('x % y =',x%y)

# Output: x // y = 3
print('x // y =',x//y)

# Output: x ** y = 50625
print('x ** y =',x**y)
```

```
x = 60
print(x) # assign values to variables
         # assign 5 to x

x += 3
print(x) # x = x + 3

x -= 5
print(x) # x = x - 5

x *= 6
print(x) # x = x * 6

x /= 3
print(x) # x = x / 3

x %= 6
print(x) # x = x % 6

x //= 1
print(x) # x = x // 1

x **= 16
print(x) # x = x ** 16
```

## while Loop - while 迴圈


```
# 試著輸入1輸出到100
i = 1
while i <= 50:
  print(i)
  i += 1
```

```
cnt=1 #this is the initial variable
while cnt < 5 :
        #inside of while loop
        print (cnt,"This is inside of while loop")
        cnt+=1
else :
        #this statement will be printed if cnt is equals to 5
        print (cnt, "This is outside of while loop")
```

```
# The else Statement
i = 1
while i < 6:
  print(i)
  i += 1
else:
  print("i is no longer less than 6")
```

```
# Python Nested while loop
line=1 #this is the initial variable
while line <= 5 :
        pos = 1
        while pos < line:
 
                #This print will add space after printing the value
                print(pos),
                
                #increment the value of pos by one
                pos += 1
        else:
                #This print will add newline after printing the value
                print(pos)
        
        #increment the value of line by one
        line += 1
```

## for loop - for 迴圈

```
word="anaconda"
for letter in word:
	print (letter)
```

```
# range()
for x in range(3):
    print("Printing:", x)
```

```
for n in range(1, 26, 1):
    print("Printing with step:", n)
```

## The break Statement -- 立刻跳出迴圈
 - With the break statement we can stop the loop even if the while condition is true:

```
i = 1
while i < 6:
  print(i)
  if i == 3:
    break
  i += 1
```

```
fruits = ["apple", "banana", "cherry"]
for x in fruits:
  print(x)
  if x == "banana":
    break
```

## The continue Statement -- 回到迴圈的起點
  - With the continue statement we can stop the current iteration, and continue with the next:






```
i = 0
while i < 6:
  i += 1
  if i == 3:
    continue
  print(i)
```

```
fruits = ["apple", "banana", "cherry"]
for x in fruits:
  if x == "banana":
    continue
  print(x)
```

## The else Statement
  - With the else statement we can run a block of code once when the condition no longer is true:

```
for x in range(6):
  print(x)
else:
  print("Finally finished!")
```

```
i = 1
while i < 6:
  print(i)
  i += 1
else:
  print("i is no longer less than 6")
```

## The pass Statement
  - The pass statement does nothing
  - It can be used when a statement is required syntactically but the program requires no action

```
while True:
    pass  # Busy-wait for keyboard interrupt (Ctrl+C)
```

```
for x in [0, 1, 2]:
  pass
```

```
# Basic 8 質數判別 -- while loop
n = int(input(''))
c = 2

while c < n:
    if n % c == 0:
        print('NO')
        break
    c += 1

if c == n:
    print('YES')
```
