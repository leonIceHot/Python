```
#@title 搜尋與選擇
```

```
poem = '''All that doth flow we cannot liquid name
Or else would fire and water be the same;
But that is liquid which is moist and wet
Fire that property can never get.
Then 'tis not cold that doth the fire put out
But 'tis the wet that makes it die, no doubt.'''

word ='the'
print(poem.find(word))
print(poem.index(word))

word2 = 'the'
print(poem.rfind(word2))
print(poem.rindex(word2))

```

```
#@title 大小寫
```

```
setup = 'a duck goes into a bar...'
print("setup     ",setup)
print("capitalize",setup.capitalize()) # 首字母大小
print("title     ",setup.title())      # 所有單字第一個字母改為大寫
print("upper     ",setup.upper())      # 所有字改為大寫
print("lower     ",setup.lower())      # 所有字改為小寫
print("swapcase  ",setup.swapcase())   # 大小寫對調
```

```
#@title 對齊方式
```

```
setup = 'a duck goes into a bar...'

print("setup",setup)
print("ljust",setup.ljust(30)) # 靠左對齊
print("rjust",setup.rjust(30)) # 靠右對齊
```

```
#@title 延續文字
```

```
# ch2_4.py +\ 延續文字

a = b = c = 10
x = a + b + c + 12
print(x)
# 續行方法1
y = a +\
    b +\
    c +\
    12
print(y)
# 續行方法2
z = ( a +     # 此處可以加上註解  
      b +
      c +
      12 )
print(z)
```
