```
# ch4_21.py
loan = eval(input("請輸入貸款金額："))
year = eval(input("請輸入年限："))
rate = eval(input("請輸入年利率："))
monthrate = rate / (12*100)             # 改成百分比以及月利率

# 計算每月還款金額
molecules = loan * monthrate
denominator = 1 - (1 / (1 + monthrate) ** (year * 12))
monthlyPay = molecules / denominator    # 每月還款金額
totalPay = monthlyPay * year * 12       # 總共還款金額

print("每月還款金額 %d" % int(monthlyPay))
print("總共還款金額 %d" % int(totalPay))

```

```
# ch4_22.py 計算正五角形的面積
# 餘角關係 tan(90°-θ) = cot(θ)
# 五邊形 https://zh.wikipedia.org/wiki/%E4%BA%94%E8%BE%B9%E5%BD%A2
# python math https://docs.python.org/zh-cn/3/library/math.html

import math

s = eval(input("請輸入正五角形邊長 : "))
area = (5 * s ** 2) / (4 * math.tan(math.pi / 5))
print("area = ", area)

```

```
# ch4_23.py 經緯度計算各城市間的距離
# math.radian() 將角度轉換成弧度

import math

r = 6371                        # 地球半徑
x1, y1 = 22.2838, 114.1731      # 香港紅磡車站經緯度
x2, y2 = 25.0452, 121.5168      # 台北車站經緯度

d = 6371*math.acos(math.sin(math.radians(x1))*math.sin(math.radians(x2))+
                   math.cos(math.radians(x1))*math.cos(math.radians(x2))*
                   math.cos(math.radians(y1-y2)))

print("distance = ", d)

```
