# Tuple

## Basic Operation

```
numbers = (1,2,3,4,5)
fruits = ('apple','banna','orange')
mixed = ('James',123)

print(f'number:{numbers}{type(numbers)}')
print(f'fruits:{fruits}{type(fruits)}')
print(f'mixed:{mixed}{type(mixed)}')
```

```
help(tuple)
```

## Indexing Tuples

```
x = (1,2,3)
print(f'x[0]:{x[0]}')
print(f'x[1]:{x[1]}')
print(f'x[2]:{x[2]}')

print(f'x[-1]:{x[-1]}')
print(f'x[-2]:{x[-2]}')
print(f'x[-3]:{x[-3]}')
```

## Tuples are immutable
  - Tuples are unchangeable , meaning that you cannot change , add , remove items 

```
x = (1,2,3)
print(f"Before:{x}")
x[0]=5
print(f"After:{x}")
```

```
a = (1,2,3,[1,2,3])
print(f'Before a : {a}')
a[3] += [4,5]
print(f'After a : {a}')
```

## Loop Tuples

```
keys = (1,2,3)
for key in keys:
    print(key)
```

```
thp = ("apple","banana","cherry")
i = 0
while i < len(thp):
    print(thp[i])
    i += 1
```

## Join Tuples

```
tuple1 = ("a","b","c")
tuple2 = (1,2,3)

tuple3 = tuple1 + tuple2
print(tuple3)

tuple4 = tuple1 * 5
print(tuple4)
```

```
t = (1,2)
print(f'Before t: {t}')
q = t
t += (3,4)
print(f'After t:{t}')
print(f'q:{q}')
```

## Packing and Unpacking Tuples

```
fruits = ("apple","banana","cherry")
#(green,yellow,red) = fruits
(yellow,green,red) = fruits
print(green)
print(yellow)
print(red)
```

```
b,*c,a = 1,2,3,4,5
print(f'a:{a}')
print(f'b:{b}')
print(f'c:{c}')
```

```
fruits = ("apple","banana","cherry","papaya","pineapple")
(green,*yellow,red) = fruits

print(green)
print(yellow)
print(red)
```

## Built-in Tuple Functions

```
tup = ('this','is','a','tuple')
print(f'len : {len(tup)}')
print(f'max : {max(tup)}')
print(f'min : {min(tup)}')
print("count",tup.count('is'))
print("index",tup.index('a'))
```

## Generator

```
example = (i/2 for i in [0,6,4,3])

for item in example:
    print(item)
```
