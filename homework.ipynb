```
import datetime

#獲得當前時間
now = datetime.datetime.now()  #->這是時間陣列格式
#轉換為指定的格式:
otherStyleTime = now.strftime("%Y/%m/%d")

def F(C):
    degree = C * 1.8 + 32
    return degree        

print("執行結果 :")

print("今天的日期為 :",otherStyleTime,end="")
C = int(input(("請輸入攝氏溫度 : \b")))
print("華氏溫度為 : %.1f" %F(C))
```

```
a,b,c,d,e,f,g,h,i,j,k,l,m,n,o= input().split()
list1 =[a,b,c,d,e,f,g,h,i,j,k,l,m,n,o]
list2= list(map(int,list1))
length = len(list2)

list3=[] # 質數
list4=[] # 非質數

for i in range(length):
    num = list2[i]
    # 質數大於 1
    if num > 1:
      # 查看因數
      for i in range(2,num):
          if (num % i) == 0:              
              list4.append(num)
              break
      else:          
          list3.append(num)         
          
    # 如果輸入的數字小於或等於 1，不是質數
    else:      
      list4.append(num)

# 限 Python 3.7
print(f"質數有 : {len(list3)}")
[print(x) for x in list3]
print(f"非質數 : {len(list4)}")
[print(y) for y in list4]
```

```
import numpy as np

list1 = []
list1 = input().split()
list2 = list(map(int,list1))
length = len(list2)

list3 = []

i = 0
while i <= length:
    num = list2[i]    
    if num >= 0:
        list3.append(num)
        i += 1
        
    elif num == -1:
        print("成績輸入結束")        
        length2 = len(list3)
        print(f"總共輸入{length2}筆資料") # 限Python3.7
        print("所有輸入成績")
        [print(x) for x in list3]        
        print("班上最高分",max(list3))
        print("班上最低分",min(list3))        
        print("班上總成績",sum(list3))
        print("全班平均分數",np.mean(list3))        
        break;       
```
